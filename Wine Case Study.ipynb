# Analyzing Wine (Red & White) Quality Datasets

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

```python
red_ds=pd.read_csv(r"C:\Users\vapa1007\Downloads\Data Analysis Case Study\winequality\winequality-red.csv", sep=";")
```

```python
white_ds=pd.read_csv(r"C:\Users\vapa1007\Downloads\Data Analysis Case Study\winequality\winequality-white.csv", sep=";")
```

```python
red_ds.head()
```

```python
white_ds.head()
```

```python
new_labels=list(red_ds.columns)
new_labels[6]='color'
red_ds.columns=new_labels
```

```python
red_ds["color"]="Red"
red_ds
```

```python
white_ds["color"]="White"
white_ds
```

```python
wine_ds=pd.merge(red_ds, white_ds,how='outer')
```

```python
wine_ds.head()
```

```python
wine_ds['total sulfur dioxide'] = wine_ds['total sulfur dioxide'].fillna(0)
```

```python
wine_ds
```

```python
wine_ds.mean()
```

```python
#Due to non-numeric columns we need to add 'numeric_only=True'

wine_ds.mean(numeric_only=True)
```

```python
wine_ds.groupby(['quality','color'],as_index=False)['pH'].mean()
```

```python
colors = ['red','white']
color_means = wine_ds.groupby('color')['quality'].mean()
color_means.plot(kind='bar',title='Average Wine Quality by Color', color=colors, alpha=.7)
plt.xlabel('Colors', fontsize=18)
plt.ylabel('Quality', fontsize=18)
```

```python
counts=wine_ds.groupby(['quality','color']).count()
counts
```

```python
# Giving an arbitary one in Y-axis as the counts above are same.

counts=wine_ds.groupby(['quality','color']).count()['pH']
counts.plot(kind='bar',title='Counts by Wine Color & Quality', color=colors, alpha=.7)
plt.xlabel('Quality & Colors', fontsize=18)
plt.ylabel('Counts', fontsize=18)
```

```python
#Since white samples were more it is hard to make fair comparision with just count.
#We will divide each count with the total count of the color to use proportion instead.

total = wine_ds.groupby('color').count()['pH']
proportion = counts/total
proportion.plot(kind='bar', title='Proportion by Wine Color & Quality', color=colors, alpha=.7)
plt.xlabel('Quality & Colors', fontsize=18)
plt.ylabel('Proportion', fontsize=18)
```

Proportion of Wine Color & Quality using Matplotlib

```python
color_counts = wine_ds.groupby(['color', 'quality']).count()['pH']
color_counts
```

```python
color_totals = wine_ds.groupby('color').count()['pH']
color_totals
```

```python
red_proportions = color_counts['Red'] / color_totals['Red']
red_proportions
```

```python
white_proportions = color_counts['White'] / color_totals['White']
white_proportions
```

```python
red_proportions['9'] = 0
red_proportions
```

```python
ind = np.arange(len(white_proportions))
width = 0.35 
```

```python
red_bars = plt.bar(ind, red_proportions, width, color='Red', alpha=.7, label='Red Wine')
white_bars = plt.bar(ind + width, white_proportions, width, color='White', alpha=.7, label='White Wine')

plt.ylabel('Proportion')
plt.xlabel('Quality')
plt.title('Proportion by Wine Color and Quality')
locations = ind + width / 2
labels = ['3', '4', '5', '6', '7', '8', '9']
plt.xticks(locations, labels)

plt.legend();
```
